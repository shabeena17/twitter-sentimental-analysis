## Sentimental Analysis of Tweets on Pepsi Advertisement

```python
# importing regular expression and stopwords package
from nltk.corpus import stopwords #install this package if this line gives you an error
from collections import Counter
from __future__ import division
import re
#function to read input file and filter stop words
def filterStopWords():
    filtered_words =[]
    stop = stopwords.words('english')
    with open('pepsi.txt','rU') as f: #reading in mode 'rU' to treat carriage return as end of line characters
        for line in f:
            line = line.decode("utf8").encode("ascii",'ignore')
            line = line.lower().rstrip()
            line = re.sub('RT\s@[A-Za-z0-9]([^\s]+)','',line) #remove twitter handle at the beginning
            line = re.sub('[-?.:!,&;\')(~|"]','',line) #remove additonal special characters
            line = re.sub('\https\/\/.*','',line) #remove all hyperlinks embedded in the tweets
            line = re.sub('#([^\s]+)','',line) #removing additional hash tags in the tweets
            line = re.sub('@([^\s]+)','',line) # remove additional tagging in the tweets           
            words = line.split(" ")
            for word in words:
                if word not in stop:
                    filtered_words.append(word)
    print filtered_words
    return filtered_words
```

### Reading positive and negative word list in the program

```python
filtered_words = filterStopWords() #invoking method to read file and create a list with all stop words eliminated
print len(filtered_words)
with open('negative-words.txt') as f2: #reading external file containing list of negative words
    neg_words = f2.read().splitlines()
with open('positive-words.txt') as f3: #reading external file containing list of positive words
    pos_words = f3.read().splitlines() 
```

### Going over every word in the tweet against pos/neg word list. Assign +1 for positive and -1 for negative word

```python
sentimentcount = 0
poscount = 0
negcount = 0
totalwordsconsidered = 0
for word in filtered_words: #iterating all the words which are not stop words
    if(word in pos_words):
        poscount +=1 #incrementing positive word count if positive word is encountered
        sentimentcount = sentimentcount + 1
    elif(word in neg_words):
        negcount +=1 #incrementing negative word count if negative word is encountered
        sentimentcount = sentimentcount - 1  
totalwordsconsidered = poscount + negcount
print 'Total Sentiment Value = ' + str(sentimentcount)
```

```python
print 'Total Positive Word Count = '+ str(poscount)
print 'Total Negative Word Count = '+ str(negcount)
print 'Total Words Considered (after filering) = '+ str(totalwordsconsidered)

percentage_negcount = (negcount / totalwordsconsidered) *100
percentage_poscount = (poscount / totalwordsconsidered)*100

print 'Percentage of Positive Word Count = '+ str(round(percentage_poscount,2)) + '%'
print 'Percentage of Negative Word Count = '+ str(round(percentage_negcount,2)) + '%'
```

### Since percentage of negative words is more than 50% of the total words considered, we conclude that the overall sentiment towards pepsi advertisement is strongly negative
